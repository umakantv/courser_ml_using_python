# Neural Network - Predicting
### Saved Parameters are loaded, Predictions are calculated

```python
import numpy as np
import scipy.io as sio
```

```python
thetas = sio.loadmat('Practice\Machine Learning\machine-learning-ex3\ex3\ex3weights.mat')
data = sio.loadmat('Practice\Machine Learning\machine-learning-ex3\ex3\ex3data1.mat')
X = data['X']
X = np.insert(arr=X, obj=0, values=1.0, axis=1)
y = data['y']
t1 = thetas['Theta1'].T
t2 = thetas['Theta2'].T
```

```python
input_layer_size = 400
hidden_layer_size = 25
num_labels = 10
```

```python
def sigmoid(Z):
    return 1/(1+1/(np.e**Z))
```

```python
q = X.dot(t1)
q = sigmoid(q)
q = np.insert(q, 0, 1.0, 1)
```

```python
p = q.dot(t2)
p = sigmoid(p)
p = np.where(p == np.amax(p, axis=1).reshape((X.shape[0], 1)) )[1]
p = p+1
p = p.reshape((X.shape[0], 1))
```

```python
acc = p[p==y].size/X.shape[0]*100
print(acc)
```
