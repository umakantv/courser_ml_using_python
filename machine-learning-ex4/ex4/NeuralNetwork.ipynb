# Neural Network

```python
import numpy as np
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

```python
# Dataset from sklearn
data= load_digits()
X = data.data
y = data.target
S = [64, 40, 10]

i = 60
x = X[i]
im = x.reshape(8, 8)
plt.gray()
plt.imshow(im)
plt.text(0, 1, y[i], bbox={'facecolor': 'black', 'pad': 10})
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='adam', hidden_layer_sizes=(30, ))
clf.fit(X, y)
```

```python
pre = clf.predict(X)
acc = pre[pre==y].size/m*100
print(acc)
clf.predict([X[i, :]])
```
